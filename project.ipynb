```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train_dataset.txt',sep = ';',header = None,names = ['Text','Emotion'])
```

```python
df['Emotion'] = df['Emotion'].map({"joy":0,"sadness":1,"anger":2,"fear":3,"love":4,"surprise":5})
```

```python
df
```

```python
df['Text'] = df['Text'].apply(lambda x : x.lower())
df['Text']
```

```python
#remove punchuation from data
import string

def remove_punc(txt):
  return txt.translate(str.maketrans('','',string.punctuation))
```

```python
df['Text'] = df['Text'].apply(remove_punc)
```

```python
#remove number from data 
def remove_numbers(txt):
    new = ""
    for i in txt:
        if not i.isdigit():
            new = new + i
    return new

df['Text'] = df['Text'].apply(remove_numbers)
```

```python
#remove emojis from data
def remove_emojis(txt):
    new = ""
    for i in txt:
        if i.isascii():
            new += i
    return new

df['Text'] = df['Text'].apply(remove_emojis)
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
```

```python
df.loc[1]['Text']
```

```python
def remove(txt):
  words = txt.split()
  cleaned = []
  for i in words:
    if not i in stop_words:
      cleaned.append(i)

  return ' '.join(cleaned)
```

```python
df['Text'] = df['Text'].apply(remove)
```

```python
df.loc[1]['Text']
```

```python
df
```

```python
df['Emotion'].value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df['Text'], df['Emotion'], test_size=0.20, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Check for Bag_Of_Words 


```python
bag_of_word_vectorizer = CountVectorizer()
X_train_bow = bag_of_word_vectorizer.fit_transform(X_train)
X_test_bow = bag_of_word_vectorizer.transform(X_test)
```

```python
logistic_model = LogisticRegression()
logistic_model.fit(X_train_bow,y_train)
```

```python
log_pred = logistic_model.predict(X_test_bow)
print(accuracy_score(y_test, log_pred))
```

# Check For TFIDFVectorizer

```python
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

```python
logistic_model = LogisticRegression(max_iter=1000)
logistic_model.fit(X_train_tfidf,y_train)
```

```python
log_pred = logistic_model.predict(X_test_tfidf)
print(accuracy_score(y_test,log_pred ))
```
